```python
class Payment():
    def __init__(self,details, amount, additive_fee=0, pct_fee=0):
        self.details= details
        self.flat_fee = additive_fee
        self.pct_fee = pct_fee
        self.amount = amount
        self.payment_type = ''
    
    def check_credentials(self):
        pass
    
    def calculate_total(self):
        pass
```

```python
class Credit_card(Payment):
    def __init__(self, details, amount, standard, additive_fee=0, pct_fee=0):
        Payment.__init__(self, details, amount, additive_fee=0, pct_fee=0)
        self.standard = standard
        self.payment_type = 'Credit card'
    
    def check_credentials(self):
        print('checking credentials...')
        for detail in self.details.keys():
            if not detail or (len(self.details[detail]) != self.standard[detail]):
                return False
        return True

    def calculate_total(self):
        if Credit_card.check_credentials(self):
            fee = self.amount * self.pct_fee + self.flat_fee
            total = self.amount + fee
            return {"success": True, "method": self.payment_type, "amount": total, "fee": fee}
        return {"success":False, "message":"Invalid input. Please check that you types your credentials correctly"}
```
