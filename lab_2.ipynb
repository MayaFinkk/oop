### Problem 1
Implement a complex number class. Overload the addition and subtraction operators to add and subtract two complex numbers.


```python
class Complex:
    def __init__(self, r, i):
        self.real = r # real part
        self.im = i # imaginary part
    
    def __add__(self, n2 : Complex):
        return Complex(self.real + n2.real, self.im + n2.im)
        
    def __sub__(self, n2 : Complex):
        return Complex(self.real - n2.real, self.im - n2.im)
        
    def __repr__(self):
        if self.im <0:
            return f'{self.real}{self.im}i'
        return f'{self.real}+{self.im}i'
```

```python
def plus_complex(n1 : Complex, n2 : Complex):
    # addition of two complex numbers
    r = n1.real + n2.real
    i = n1.im + n2.im
    result = Complex(r, i)
    print(n1, ' + ', n2, " = ", result, sep='')
    return result
```

### Problem 2
Create a class `Person` with `name`, `age` and `gender` as its attributes. Overload the `>` operator to compare two persons based on their age, such that `p1 > p2` returns `True` if the age of `p1` is greater than `p2` and `False` otherwise.

```python
class Person:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def __gt__(self, dude):
        if self.age > dude.age:
            return True
        return False
    
    def __lt__(self, dude):
        if self.age < dude.age:
            return True
        return False
    
```

### Problem 1
Create a class `Vehicle` that has a method `drive` that returns the string "Driving a vehicle". Create two subclasses, `Car` and `Bicycle`, that override this method to return "Driving a car" and "Riding a bicycle", respectively.

```python
class Vehicle:
    def __init__(self):
        self.type = 'vehicle'
        
    def drive(self):
        return f'Driving a {self.type}'
    
class Car(Vehicle):
    def __init__(self):
        self.type = 'car'
        
        
class Bicycle(Vehicle):
    def __init__(self):
        self.type = 'bicycle'
        
```
